```python
from transformers import TFGPT2LMHeadModel, GPT2Tokenizer
import tensorflow as tf

# Load pre-trained tokenizer and model
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = TFGPT2LMHeadModel.from_pretrained("gpt2")
```

```python
def generate_story_tf(prompt, max_length=100, temperature=0.9, top_k=50):
    # Encode the input prompt
    input_ids = tokenizer.encode(prompt, return_tensors='tf')

    # Generate output using sampling
    output = model.generate(
        input_ids,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        do_sample=True,
        no_repeat_ngram_size=2,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode the output tokens into text
    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
prompt = "Once upon a time, there was a robot doctor living in Dhaka city"
story = generate_story_tf(prompt, max_length=150)
print(story)
```
