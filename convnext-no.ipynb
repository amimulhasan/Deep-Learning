```python
# ======================================================
# Step 0: Imports
# ======================================================
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import ResNet50
from PIL import Image
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# ======================================================
# Step 1: ViT Custom Layers
# ======================================================
class Patches(layers.Layer):
    def __init__(self, patch_size, **kwargs):
        super().__init__(**kwargs)
        self.patch_size = patch_size
    def call(self, images):
        batch_size = tf.shape(images)[0]
        patches = tf.image.extract_patches(
            images=images,
            sizes=[1,self.patch_size,self.patch_size,1],
            strides=[1,self.patch_size,self.patch_size,1],
            rates=[1,1,1,1],
            padding="VALID"
        )
        patches = tf.reshape(patches, [batch_size, -1, patches.shape[-1]])
        return patches
    def get_config(self):
        return {"patch_size": self.patch_size}

class PatchEncoder(layers.Layer):
    def __init__(self, num_patches, projection_dim, **kwargs):
        super().__init__(**kwargs)
        self.num_patches = num_patches
        self.projection = layers.Dense(projection_dim)
        self.position_embedding = layers.Embedding(num_patches, projection_dim)
    def call(self, patches):
        positions = tf.range(start=0, limit=self.num_patches)
        return self.projection(patches) + self.position_embedding(positions)
    def get_config(self):
        return {"num_patches": self.num_patches, "projection_dim": self.projection.units}

# ======================================================
# Step 2: Dataset Paths & Classes
# ======================================================
base_path = "/kaggle/input/datasets/amimulahasanrofik/alz-b-1100/alzheimer_new_11/alzheimer_new"

alzheimer_classes = ["MildDemented","Moderate Dementia","NonDemented","Very mild Dementia"]
tumor_classes = ["glioma","meningioma","notumor","pituitary"]
all_classes = alzheimer_classes + tumor_classes

# Stage-1 mapping: Alzheimer vs Tumor
disease_mapping = {c:0 for c in alzheimer_classes}
disease_mapping.update({c:1 for c in tumor_classes})

# ======================================================
# Step 3: Image Loader
# ======================================================
IMG_SIZE = (128,128)

def load_images(classes, mapping=None):
    X, y = [], []
    for cls in classes:
        cls_path = os.path.join(base_path, cls)
        for img_name in os.listdir(cls_path):
            img_path = os.path.join(cls_path, img_name)
            try:
                img = Image.open(img_path).convert("RGB")
                img = img.resize(IMG_SIZE)
                img_array = np.array(img, dtype=np.float32)/255.0
                X.append(img_array)
                if mapping:
                    y.append(mapping[cls])
                else:
                    y.append(all_classes.index(cls))
            except:
                continue
    return np.array(X), np.array(y)

# ======================================================
# Step 4: Stage-1 Dataset
# ======================================================
X_stage1, y_stage1 = load_images(all_classes, disease_mapping)
X_train1, X_val1, y_train1, y_val1 = train_test_split(
    X_stage1, y_stage1, test_size=0.2, stratify=y_stage1, random_state=42
)

# ======================================================
# Step 5: Stage-1 Model (ResNet50 + ViT + GRU)
# ======================================================
def build_stage1_model(input_shape=(128,128,3), patch_size=16, projection_dim=64, transformer_layers=2, num_heads=4):
    inputs = layers.Input(shape=input_shape)

    # ----- ResNet50 Branch -----
    base_model = ResNet50(weights='imagenet', include_top=False, input_tensor=inputs)
    base_model.trainable=False
    x_cnn = layers.GlobalAveragePooling2D()(base_model.output)

    # ----- ViT + GRU Branch -----
    x_vit = inputs
    patches = Patches(patch_size)(x_vit)
    num_patches = (input_shape[0]//patch_size)*(input_shape[1]//patch_size)
    encoded_patches = PatchEncoder(num_patches, projection_dim)(patches)

    for _ in range(transformer_layers):
        x1 = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
        attn = layers.MultiHeadAttention(num_heads=num_heads, key_dim=projection_dim)(x1,x1)
        encoded_patches = layers.Add()([encoded_patches, attn])

    x_vit = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
    x_vit = layers.Flatten()(x_vit)
    x_vit = layers.Reshape((-1, x_vit.shape[-1]))(x_vit)
    x_vit = layers.GRU(128)(x_vit)

    # ----- Concatenate -----
    x = layers.concatenate([x_cnn, x_vit])
    x = layers.Dense(128, activation='relu')(x)
    outputs = layers.Dense(2, activation='softmax')(x) # Stage-1: Alzheimer vs Tumor

    model = models.Model(inputs, outputs)
    return model

model_stage1 = build_stage1_model()
model_stage1.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_stage1.summary()

# ======================================================
# Step 6: Train Stage-1
# ======================================================
history1 = model_stage1.fit(
    X_train1, y_train1,
    validation_data=(X_val1, y_val1),
    epochs=20,
    batch_size=32
)

# ======================================================
# Step 7: Stage-2 Datasets
# ======================================================
# Alzheimer 4-class
X_alz, y_alz = load_images(alzheimer_classes)
X_train_alz, X_val_alz, y_train_alz, y_val_alz = train_test_split(
    X_alz, y_alz, test_size=0.2, stratify=y_alz, random_state=42
)

# Tumor 4-class
X_tumor, y_tumor = load_images(tumor_classes)
X_train_tumor, X_val_tumor, y_train_tumor, y_val_tumor = train_test_split(
    X_tumor, y_tumor, test_size=0.2, stratify=y_tumor, random_state=42
)

# ======================================================
# Step 8: Stage-2 Models
# ======================================================
def build_stage2_model(num_classes):
    inputs = layers.Input(shape=(128,128,3))
    base_model = ResNet50(weights='imagenet', include_top=False, input_tensor=inputs)
    base_model.trainable=False
    x_cnn = layers.GlobalAveragePooling2D()(base_model.output)

    x_vit = inputs
    patches = Patches(16)(x_vit)
    num_patches = (128//16)*(128//16)
    encoded_patches = PatchEncoder(num_patches, 64)(patches)
    x_vit = layers.LayerNormalization(epsilon=1e-6)(encoded_patches)
    x_vit = layers.Flatten()(x_vit)
    x_vit = layers.Reshape((-1, x_vit.shape[-1]))(x_vit)
    x_vit = layers.GRU(128)(x_vit)

    x = layers.concatenate([x_cnn, x_vit])
    x = layers.Dense(128, activation='relu')(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    return models.Model(inputs, outputs)

model_alz = build_stage2_model(4)
model_alz.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model_tumor = build_stage2_model(4)
model_tumor.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# ======================================================
# Step 9: Train Stage-2 Models
# ======================================================
history_alz = model_alz.fit(X_train_alz, y_train_alz, validation_data=(X_val_alz, y_val_alz), epochs=25, batch_size=32)
history_tumor = model_tumor.fit(X_train_tumor, y_train_tumor, validation_data=(X_val_tumor, y_val_tumor), epochs=25, batch_size=32)

# ======================================================
# Step 10: Save Models
# ======================================================
model_stage1.save("stage1_resnet50_vit_gru.keras")
model_alz.save("stage2_alz_resnet50_vit_gru.keras")
model_tumor.save("stage2_tumor_resnet50_vit_gru.keras")

# ======================================================
# Step 11: Hierarchical Prediction Function
# ======================================================
def hierarchical_predict(img_path):
    img = Image.open(img_path).convert("RGB").resize((128,128))
    img_array = np.expand_dims(np.array(img)/255.0, axis=0)

    stage1_pred = np.argmax(model_stage1.predict(img_array, verbose=0))
    if stage1_pred==0:  # Alzheimer
        pred_idx = np.argmax(model_alz.predict(img_array, verbose=0))
        return alzheimer_classes[pred_idx]
    else:  # Tumor
        pred_idx = np.argmax(model_tumor.predict(img_array, verbose=0))
        return tumor_classes[pred_idx]

# ======================================================
# Step 12: Test on Unknown Dataset
# ======================================================
unknown_path = "/kaggle/input/datasets/amimulahasanrofik/bra-alz-8/alzheimer_new"  # Change to your unknown dataset path
y_true, y_pred = [], []

for cls in all_classes:
    cls_folder = os.path.join(unknown_path, cls)
    for img_name in os.listdir(cls_folder):
        img_path = os.path.join(cls_folder, img_name)
        try:
            pred = hierarchical_predict(img_path)
            y_pred.append(pred)
            y_true.append(cls)
        except:
            continue

# Metrics
print(classification_report(y_true, y_pred, digits=4))
cm = confusion_matrix(y_true, y_pred, labels=all_classes)
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt="d", xticklabels=all_classes, yticklabels=all_classes)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Unknown Dataset Confusion Matrix")
plt.show()
```
