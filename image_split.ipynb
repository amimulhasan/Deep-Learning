```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io

# Step 1: Load the image
image = io.imread('/content/rft.jpg')

# Ensure the image is 512x512
if image.shape[0] != 512 or image.shape[1] != 512:
    raise ValueError(f"Expected image of size 512x512, but got {image.shape[0]}x{image.shape[1]}")

# Step 2: Define patch size and split the image into 9 patches (3x3 grid)
patch_size = 171  # Each patch size (512 / 3)

# Step 3: Split the image manually into 9 patches
patches = []
for i in range(3):
    for j in range(3):
        patch = image[i*patch_size:(i+1)*patch_size, j*patch_size:(j+1)*patch_size]
        patches.append(patch)

# Step 4: Visualize the 9 patches
fig, ax = plt.subplots(3, 3, figsize=(9, 9))  # 3x3 grid for 9 patches
for i in range(3):
    for j in range(3):
        ax[i, j].imshow(patches[i*3 + j], cmap='gray')  # Show each patch
        ax[i, j].axis('off')  # Hide axis
plt.show()
```
